```python
#Reading the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv("Superstore_USA.csv", encoding='latin')
#printing the dataset
print(dataset)
 
 
```

```python
#this identify the missing values from the datset
dataset.isnull().sum()
```

```python
#now we will fill mean value inplace of missing values because .......
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace=True)
```

```python
#now again checking the any missing value in dataset
dataset.isnull().sum()
```

Analysis based on Order Priority
